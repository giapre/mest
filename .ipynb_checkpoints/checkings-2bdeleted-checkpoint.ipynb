```python
import os
import pandas as pd
import mest_function
import numpy as np
from scipy.optimize import fsolve
import random
```

```python
all_the_deltas, _ = mest_function.f1_find_the_delta()

X0 = np.arange(-3.02,-2.07,0.3)
for j, jj in enumerate(X0):
    X0[j] = round(jj,2)
X0 = X0[::-1]

startingx1 = []
startingz = []
for m in X0:
    f1z, f1x1 = fsolve(mest_function.s1_inters_bif_line, (3, -1), m)
    startingz.append(f1z)
    startingx1.append(f1x1)
    
data = {'x0': X0, 'x1': startingx1, 'z': startingz}
startings = pd.DataFrame(data)
#startings.loc[0]['x1'] = startings.loc[1]['x1']
#startings.loc[0]['z'] = startings.loc[1]['z']
startings
```

```python
Is = np.arange(0.05, 1.9, 0.3)
for j, jj in enumerate(Is):
    Is[j] = round(jj,2)
```

```python
steps = 500
timesteps = np.arange(100,2050,steps)
timesteps
```

```python
# Data
data_path = '/home/prior/gp_home/mest/10-pat-data/10-pat-connect'
```

```python
for patient in ['connectivity.vep.id001_bt.zip', 'connectivity.vep.id006_mr.zip']: #os.listdir(data_path):
    
    patname = patient.split('.')[2]
    
    
    #Pastduration and dataframe reset at every new patient
    pastduration = [100]*len(Is)
    A = np.ones((len(Is),len(timesteps)))
    df = pd.DataFrame(A, index = Is, columns = timesteps)
    
    for x0 in X0: 
        print('\n Vado con ',patname, ', x0 = ', x0, '\n')
    
        for i, ii in enumerate(Is): 
            seizing = False
            duration = pastduration[i]

            while(seizing==False): 

                initx1 = round(float(startings.loc[startings['x0'] == x0]['x1']), 4)
                initz = round(float(startings.loc[startings['x0'] == x0]['z']), 4)
                
                print('Initial conditions are ', initx1, initz)
                print(ii, duration)
                
                if random.randint(0,3)%2 == 0:
                    print('Ha iniziato a I: ', ii, 'e durata: ', duration, '\n')
                    seizing = True
                    if duration > steps:
                        pastduration[i] = duration - steps
                else:
                    df.loc[ii][duration] = 0.
                    duration = duration + steps

                if duration > 2000:
                    print('Non è mai partito! \n')
                    seizing = True

        print(df)
        #df.to_csv(f'results/10-pat-results/{patname}.{x0}.csv')

```
