```python
X = []
Y = []
Z = []
for tt in X0:
    csvfile = f'{tt}.csv'
    dfan = pd.read_csv(csvfile, index_col=0)
    for c in dfan.index:
        for cc in dfan.columns:
            if dfan.loc[c][cc] == 1.0:
                X.append(c) 
                Y.append(int(cc))
                Z.append(tt)
```

```python
from scipy.interpolate import CubicSpline
multsss = []
star = []
for tt in X0:
    csvfile = f'{tt}.csv'
    dfan = pd.read_csv(csvfile, index_col=0)
    X = []
    Y = []
    Z = []
    mult = []
    for c in dfan.index:
        alreadyfound = False
        for cc in dfan.columns:
            if alreadyfound == False:
                if dfan.loc[c][cc] == 1.0:
                    X.append(c) 
                    Y.append(int(cc))
                    mult.append(c*int(cc))
                    alreadyfound = True

# 300 represents number of points to make between T.min and T.max 
    multsss.append(np.mean(mult))
    star.append(np.std(mult))
    plt.xlabel('Stimulus intensity in mA')
    plt.ylabel('Stimulus duration')
    plt.plot(X,Y)
multsss
```
